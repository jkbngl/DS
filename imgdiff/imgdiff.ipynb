```python
from PIL import Image, ImageChops
```

```python
im1 = Image.open("gear_flat.png")
im2 = Image.open("gear_high.png")

diff = ImageChops.difference(im2, im1)


diff.getbbox()

diff_int_f_vs_h = diff.getbbox()[0] * diff.getbbox()[1] * diff.getbbox()[2] * diff.getbbox()[3]
diff_int_f_vs_h

diff
```

```python
im1 = Image.open("gear_flat.png")
im2 = Image.open("gear_other.png")

diff = ImageChops.difference(im2, im1)

diff.getbbox()

diff_int_f_vs_o = diff.getbbox()[0] * diff.getbbox()[1] * diff.getbbox()[2] * diff.getbbox()[3]
diff_int_f_vs_o

diff
```

```python
im1 = Image.open("gear_flat.png")
im2 = Image.open("gear_very_flat.png")

diff = ImageChops.difference(im2, im1)

diff.getbbox()

diff_int_f_vs_vf = diff.getbbox()[0] * diff.getbbox()[1] * diff.getbbox()[2] * diff.getbbox()[3]
diff_int_f_vs_vf

diff
```

```python
all_diffs = [diff_int_f_vs_h, diff_int_f_vs_o, diff_int_f_vs_vf]
```

```python
diff_int_f_vs_h_norm = (diff_int_f_vs_h - min(all_diffs)) / (max(all_diffs) - min(all_diffs))
diff_int_f_vs_o_norm = (diff_int_f_vs_o - min(all_diffs)) / (max(all_diffs) - min(all_diffs))
diff_int_f_vs_vf_norm = (diff_int_f_vs_vf - min(all_diffs)) / (max(all_diffs) - min(all_diffs))

print(f"diff_int_f_vs_h_norm: {diff_int_f_vs_h_norm}, diff_int_f_vs_o_norm: {diff_int_f_vs_o_norm}, diff_int_f_vs_vf_norm: {diff_int_f_vs_vf_norm}")
```
