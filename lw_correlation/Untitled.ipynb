```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt

#import sys
#!python3 -m pip3 install plotly
#import plotly.graph_objects as go
```

```python
df = pd.read_json('data_100.json')
df["key"] = df["SITE_FK"].astype(str) + df["DDATE"]

dfNumeric = df[['VALUE', 'key', 'ABBREVIATION']]


#dfNumeric = dfNumeric.pivot(columns='ABBREVIATION')['VALUE']
dfNumeric = df.pivot(index='key', columns='ABBREVIATION')['VALUE']

dfNumeric = dfNumeric.drop(columns=['OEEFRGG', 'OEEFRGR'])


dfNumeric.head(10)
```

```python
dfCorr = dfNumeric.corr(method='pearson')
```

```python
from matplotlib.pyplot import figure

figure(num=None, figsize=(16, 12), dpi=80, facecolor='w', edgecolor='k')
mask = np.zeros_like(dfCorr)
mask[np.triu_indices_from(mask)] = True

ax = sns.heatmap(dfCorr, mask=mask, linewidth=0.5, square=True, cmap="RdYlGn")

plt.show()
```
